```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import balanced_accuracy_score
from catboost import CatBoostClassifier
```

```python
X_train=pd.read_csv('train_x.csv')
X_test=pd.read_csv('test_x.csv')
y_train=pd.read_csv('train_y.csv')
```

```python
X_train=X_train.loc[y_train.dropna().index]
```

```python
y_train=y_train.dropna()
```

```python
y_train
```

```python
nafill=dict(X_train.describe().T['mean'])
nafill['Age']=0.0
nafill['Loan type']='A' #this is randomly set
nafill['Occupation type']='Y' #this is randomly set
nafill
```

```python
X_train=X_train.fillna(value=nafill)
```

```python
X_train 
```

```python
X_train.describe()
```

```python
X_train=pd.get_dummies(X_train,columns=['Loan type','Occupation type']).drop(columns='ID')
```

```python
X_train
```

```python
y_train=y_train.loc[:,'Label']
```

```python
y_train
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,test_size=0.2, random_state=0)
```

## Gradient Boosting Classifier

```python
clf = GradientBoostingClassifier(learning_rate=0.1,random_state=0)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_val)
```

```python
print('training balanced accuracy:')
balanced_accuracy_score(y_train,clf.predict(X_train))
```

```python
print('validation balanced accuracy:')
balanced_accuracy_score(y_val,clf.predict(X_val))
```

```python
accuracy_score(y_train,clf.predict(X_train))
```

```python
accuracy_score(y_val,clf.predict(X_val))
```

```python
confusion_matrix(y_val,clf.predict(X_val))
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=SVC()
clf.fit(X_train,y_train)
```

```python
print('training balanced accuracy:',balanced_accuracy_score(y_train,clf.predict(X_train)))
print('Vaidation balanced accuracy:',balanced_accuracy_score(y_val,clf.predict(X_val)))
```

```python
clf = RandomForestClassifier(random_state=0,class_weight='balanced',criterion='entropy',max_depth=20,n_jobs=-1)
clf.fit(X_train,y_train)
```

```python
print('training balanced accuracy:',balanced_accuracy_score(y_train,clf.predict(X_train)))
print('Vaidation balanced accuracy:',balanced_accuracy_score(y_val,clf.predict(X_val)))
```
